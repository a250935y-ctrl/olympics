 「人類の『形』は変わったか？」— 体格の進化と専門化

100年前のメダリストと現代のメダリストは、同じ人間とは思えないほど体格が異なっているかもしれません。



切り口: 競技ごとに「身長・体重・BMI」の推移を可視化します。

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#読み込む
df = pd.read_csv('olympics.csv').dropna(subset=['Height', 'Weight'])
#データの選出(体操とバスケ)
target = df[df['Year'].isin([1896, 2016]) & df['Sport'].isin(['Gymnastics', 'Basketball'])]
#描写する
plt.figure(figsize=(10, 6))
sns.scatterplot(data=target, x='Height', y='Weight', hue='Sport', style='Year', s=100)

plt.title('Olympic Body Evolution (1896 vs 2016)')
plt.show()
```
